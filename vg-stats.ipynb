### No. 5 For North American video game sales, what's the median?

### No. 6 For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?
### Answer:

```python
(df['NA_Sales'].head(1) - df['NA_Sales'].mean()) / df['NA_Sales'].std()
```

### No. 7 The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
df[df['Platform'] == 'Wii'][['Platform', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].groupby(df.Platform).mean()
```

```python
df[df['Platform'] != 'Wii'][['Platform', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].groupby(df.Platform).mean()
```
