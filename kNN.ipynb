### 4. Split the data into training set and test set. Make a test set size 0.3 and random seed 123 (so the results are consistent for all students)

```python
# Return new object with labels in requested axis removed.
X, Y = df.drop(columns=['TARGET CLASS']), df['TARGET CLASS']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=123)
```

### 5. Standartize the variables - scale them to zero mean and unit variance. Do not pass the "TARGET CLASS" column to the scaler!

```python
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

### 6. Check that the features are indeed scaled

```python
plt.plot(X_train, Y_train)
```

```python
plt.plot(X, Y)
```

### 7. Fit the kNN to the training data. Use k = 1.

```python
n_neighbors = 1
h = .02
weights = 'uniform'
clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights)
clf.fit(X_train, Y_train)
```

### 8. Check the classifier performance

```python
pred = clf.predict(X_test)
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
```
