```python
import pandas as pd
import numpy as np
from datetime import date
```

```python
df = pd.read_csv('./station.csv')
df.head()
```

```python
dtp = pd.read_csv('./trip.csv', skiprows=[50793]).drop_duplicates()
dtp.head()
```

```python
dw = pd.read_csv('./weather.csv')
dw.head()
```

### What is the average trip duration for a borrowed bicycle?
### Answer: 1202.61

```python
pd.DataFrame(dtp['tripduration']).mean()
```

### What's the most common age of a bicycle-sharer?
### Answer: 31

```python
pd.DataFrame(dtp['birthyear']).groupby(dtp.birthyear).count().sort_values('birthyear', ascending=False).head(5)
```

### Given all the weather data here, find the average precipitation per month, and the median precipitation.

```python
dw['Date'] = pd.to_datetime(dw['Date'], errors ='coerce')
dw['MONTH'] = dw['Date'].dt.month
dw['Precipitation_In'].groupby(dw.MONTH).mean()
```

### Median

```python
dw['Precipitation_In'].groupby(dw.MONTH).median()
```
